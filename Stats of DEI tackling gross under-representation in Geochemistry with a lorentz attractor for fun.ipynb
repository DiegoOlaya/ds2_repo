# Now lets see if these stats can create a chaotic trajectory!! (fun little extra)


```python
def LorentzEquations(t, S, sigma, r, b):
    x, y, z = S #define a vector space
    return [sigma * (y - x), (r * x) - y - (x * z),(x *y) - (b * z)] #returns 3 non-linear differential equations

#Specify intital conditions

x0 = 0
y0 = 1
z0 = 1

#Specify paramters obtained from the DEI and Science word count

sigma = pf_m_statsitcs * 100 #rescaled to reduce computational time

r = pf_f_statsitcs * 100 

b = pf_o_statsitcs 

#specify a timeperiod, since these websites 2019-2021 we will then make 2019 = 0, and 2020 = 100
t0 = 0
tf = 1000

#we will now solve the non-linear differential equation numerically

fullsol = integrate.solve_ivp(LorentzEquations,[t0,tf], [x0,y0,z0],
args=(sigma,r,b),t_eval=np.linspace(t0,tf,10000), dense_output=True)
```

```python
plt.plot(fullsol.y[0,:], fullsol.y[2,:])
plt.xlabel('$x(t)$')
plt.ylabel('$z(t)$')
plt.show()
```
